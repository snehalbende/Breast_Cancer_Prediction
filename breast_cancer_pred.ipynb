```python
data = data.drop('id' , axis = 1)
```

```python
cor_matrix = data.corr().abs()
```

```python
# dropping highly correlated values to get rid of multicolinearity 
up_tr = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
```

```python
# checking which columns to drop
to_drop = [column for column in up_tr.columns if any(up_tr[column] > 0.90)]
print(); print(to_drop)
```

```python
data = data.drop(['perimeter_mean', 'area_mean', 'concave points_mean', 'perimeter_se', 'area_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'concave points_worst'] , axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
data['Diagnosis'] = labelencoder.fit_transform(data['diagnosis'])
data
```

```python
X = data.drop(['Diagnosis','diagnosis'], axis = 1)
y = data['Diagnosis']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',C=1.0,random_state=0)
svm.fit(X_train,y_train)
pred3 = svm.predict(X_test)
print(confusion_matrix(y_test, pred3))
print(classification_report(y_test, pred3))
```
